### Train test split

Splitting data to train and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('loan_repaid', axis = 1), df['loan_repaid'], test_size= 0.2, random_state=101)
```

### Normalizing

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
X_train = minmax.fit_transform(X_train)
X_test = minmax.transform(X_test)
```

## 4. Create Model

### 4.1 Models with imbalanced data

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(78, activation = 'relu'))

model.add(Dense(39, activation = 'relu'))

model.add(Dense(19, activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))

# to avoid overfitting
# stopping model if the value of the function does not improve for the next 3 epochs.
es_callback = EarlyStopping(monitor='val_loss', patience=3)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x=X_train, 
          y=y_train, 
          batch_size=256,
          epochs=200,
          validation_data=(X_test, y_test), 
          callbacks=[es_callback])
```

```python
plt.plot(model.history.history['loss'], label = 'train')
plt.plot(model.history.history['val_loss'], label = 'test')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
accuracy = model.evaluate(X_test, y_test)[1]
print('Accuracy: ',accuracy)
```

```python
from sklearn.metrics import classification_report
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))
```

Because we have unbalanced data we should look at f1 score. Model is better in predicting fully paid loan.

### Model with dropout layers

```python
model1 = Sequential()


# input layer
model1.add(Dense(78,  activation='relu'))
model1.add(Dropout(0.2))

# hidden layer
model1.add(Dense(39, activation='relu'))
model1.add(Dropout(0.2))

# hidden layer
model1.add(Dense(19, activation='relu'))
model1.add(Dropout(0.2))

# output layer
model1.add(Dense(units=1,activation='sigmoid'))


model1.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model1.fit(x=X_train, 
          y=y_train, 
          epochs=50,
          batch_size=256,
          validation_data=(X_test, y_test), 
          )
```

```python
plt.plot(model1.history.history['loss'], label = 'train')
plt.plot(model1.history.history['val_loss'], label = 'test')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
model1.evaluate(X_test, y_test)[1]
```

```python
from sklearn.metrics import classification_report
y_pred1 = (model1.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred1))
```

Results are almost the same.

### 4.2 Models with balanced data

```python
X_train_1 = X_train.copy()
y_train_1 = y_train.copy()
```

```python
y_train_1.value_counts()
```

#### under sampling

Balancing data by undersampling ( keeping all of the data in the minority class and decreasing the size of the majority class).

```python
from imblearn.datasets import make_imbalance
```

```python
X_train_1, y_train_1 = make_imbalance(X_train_1, y_train_1, sampling_strategy={0: 6000, 1: 6000},random_state=14)
```

```python
model2 = Sequential()

model2.add(Dense(78, activation = 'relu'))

model2.add(Dense(39, activation = 'relu'))

#model2.add(Dense(19, activation = 'relu'))

model2.add(Dense(1, activation = 'sigmoid'))


model2.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model2.fit(x=X_train_1, 
          y=y_train_1, 
          epochs=50,
          batch_size=256,
          validation_data=(X_test, y_test), 
           callbacks=[es_callback]
          )
```

```python
plt.plot(model2.history.history['loss'], label = 'train')
plt.plot(model2.history.history['val_loss'], label = 'test')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report
y_pred2 = (model2.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred2))
```

Accuracy and f1-score are worst than in previous model with imbalanced data. It can be caused by smaller size of training set.

#### SMOTE

Second method of balancing data with oversampling.

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state = 14)
X_train_3, y_train_3 = smote.fit_resample(X_train, y_train)
```

```python
y_train_3.value_counts()
```

```python
model3 = Sequential()

model3.add(Dense(78, activation = 'relu'))

model3.add(Dense(39, activation = 'relu'))

model3.add(Dense(1, activation = 'sigmoid'))

model3.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model3.fit(x=X_train_3, 
          y=y_train_3, 
          epochs=50,
          batch_size=64,
          validation_data=(X_test, y_test),
           callbacks=[es_callback]
          )
```

```python
plt.plot(model3.history.history['loss'], label = 'train')
plt.plot(model3.history.history['val_loss'], label = 'test')
plt.ylim(0,1)
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report
y_pred3 = (model3.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred3))
```

Accuracy is better than in model with undersampling data but worst than with first model with imbalanced data.
